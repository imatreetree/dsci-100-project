```r
set.seed(420)

red_split <- initial_split(red_data, prop = 0.75, strata = quality)
red_train <- training(red_split)
red_test <- testing(red_split)

red_recipe <- recipe(quality ~ alcohol, data = red_train) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors())

knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) %>%
    set_engine("kknn") %>%
    set_mode("regression")

red_vfold <- vfold_cv(red_train, v = 5, strata = quality)

n_vals <- tibble(neighbors = seq(1, 400))

# Finds minimum RMSE

red_min <- workflow() %>%
    add_recipe(red_recipe) %>%
    add_model(knn_spec) %>%
    tune_grid(resamples = red_vfold, grid = n_vals) %>%
    collect_metrics() %>%
    filter(.metric == "rmse") %>%
    arrange(mean) %>%
    slice(1) %>%
    pull(neighbors)

red_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = red_min) %>%
    set_engine("kknn") %>%
    set_mode("regression")

red_fit <- workflow() %>%
    add_recipe(red_recipe) %>%
    add_model(red_spec) %>%
    fit(red_train)

options(repr.plot.width = 8, repr.plot.height = 8)

red_predict <- red_fit %>%
    predict(red_train) %>%
    bind_cols(red_train) %>%
    ggplot(aes(x = alcohol, y = quality)) +
    geom_point() +
    geom_line(mapping = aes(x = alcohol, .pred)) +
    xlab("Alcohol Content") +
    ylab("Quality of Wine") +
    ggtitle("Quality of Wine Based on Alcohol Content") +
    theme(text = element_text(size = 20))

red_predict
```
